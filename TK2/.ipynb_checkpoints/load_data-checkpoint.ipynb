```python
import pandas as pd
import numpy as np

df = pd.read_csv('data_covid_jakbar.csv')

# 2. 'Konfirmasi_Sembuh'
df['Konfirmasi_Sembuh'] = df['Konfirmasi_Positif'] - (df['Konfirmasi_Meninggal'] + df['Konfirmasi_Isolasi_Mandiri'])

print("Ready for Model:")
print(df)
```

```python
from sklearn.linear_model import LinearRegression

# 1. variabel independen (X) and dependen (y)
# X 2D [[...]]
X = df[['Konfirmasi_Positif']] 
y = df['Konfirmasi_Sembuh']

model = LinearRegression()

# 3. Training using fit
model.fit(X, y)

print("OK")
```

```python
# 1. GET value koefisien (m)
koefisien = model.coef_[0]

# 2. GET value konstanta/intercept (c)
konstanta = model.intercept_

print(f"Koefisien (m): {koefisien:.4f}")
print(f"Konstanta (c): {konstanta:.4f}")
print("---------------------------------------------------------")
print(f"Formula Regression: Sembuh = {koefisien:.4f} * (Positif) + {konstanta:.4f}")
```

```python
from sklearn.metrics import mean_squared_error

# 1. Model Prediction
y_prediksi = model.predict(X)

# 2. Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y, y_prediksi)

# 3. Calculate RMSE dengan sqrt MSE
rmse = np.sqrt(mse)

print(f"Nilai Mean Squared Error (MSE): {mse:.2f}")
print(f"Nilai Root Mean Squared Error (RMSE): {rmse:.2f} ")
```

```python
# Calculate value R-squared from model
r2_score = model.score(X, y)

print(f"Akurasi model (R-squared): {r2_score:.4f}")
print(f"Capability {r2_score:.2%} dari variasi data")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
# Create scatter plot from data asli
sns.scatterplot(x='Konfirmasi_Positif', y='Konfirmasi_Sembuh', data=df, s=150, label='Data Aktual per Kelurahan')
# Plot garis hasil prediksi regresi
plt.plot(X, y_prediksi, color='red', linewidth=3, label='Garis Regresi Linear')
plt.title('Hubungan Konfirmasi Positif vs. Sembuh di Jakarta Barat')
plt.xlabel('Jumlah Konfirmasi Positif')
plt.ylabel('Jumlah Konfirmasi Sembuh')
plt.legend()
plt.grid(True)
plt.show()
```
